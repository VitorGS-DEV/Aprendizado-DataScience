```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# %matplotlib notebook
```

```python
df = pd.read_csv('sinasc_RO_2019.csv')
df
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
df_graf = df[['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']]
df_graf
```

```python
plt.rc('figure', figsize=(10, 8))
fig1, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0, 0], x='APGAR1', data=df_graf)
sns.countplot(ax = axes[0, 1], x='APGAR5', data=df_graf)
sns.countplot(ax = axes[1, 0], x='SEXO', data=df_graf)
#sns.countplot(ax = axes[1, 0], x='GESTACAO', data=df_graf)
sns.countplot(ax = axes[1, 1], x='ESTCIVMAE', data=df_graf)
axes[1,1].set_xlabel('Estado civil da mãe')

plt.subplots_adjust(wspace=0.4, hspace=0.2)
```

```python
plt.close('all')
sns.countplot(x='GESTACAO', data=df_graf)
plt.xticks(rotation=30)
%matplotlib inline
plt.gcf().set_size_inches(6, 4)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
sns.barplot(x='APGAR5', y= 'CONSULTAS' , data=df)
#plt.xticks(rotation=30)

plt.gcf().set_size_inches(6, 4)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(10, 8))
fig1, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0, 0], x='SEXO', data=df)
sns.countplot(ax = axes[0, 1], x='GESTACAO', data=df)
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=df)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=df)
axes[1,0].set_xlabel('Estado civil da mãe')
axes[0,1].tick_params(axis='x', rotation=30 )
plt.subplots_adjust(wspace=0.4, hspace=0.4)
```
